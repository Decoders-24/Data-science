```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(8, 5))
sns.countplot(data=Columns_B_Category, y='Category',
              order=Columns_B_Category['Category'].value_counts().index, palette='viridis')

# Adding labels and title
plt.xlabel('Count')
plt.ylabel('Category')
plt.title('Category Value Counts')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


value_counts = Columns_B_Category['Product'].value_counts()

plt.figure(figsize=(10, 6))
value_counts.plot(kind='bar')

plt.xlabel('Product Category')
plt.ylabel('Number of Products')
plt.title('Product Count by Category')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Horizontal Bar plot

plt.figure(figsize=(8,4))
value_counts.plot(kind='barh')

plt.xlabel('Number of Products')
plt.ylabel('Product Category')
plt.title('Product Count by Category')

plt.tight_layout()
plt.show()
```

```python
#Pi Charts
plt.figure(figsize=(8 ,8))
value_counts.plot(kind='pie',autopct='%1.1f%%', startangle=90,colors=sns.color_palette('viridis'))

plt.title('Product Distribution by Category')

plt.tight_layout()
plt.show()
```

```python
# Line plot

plt.figure(figsize=(10, 6))
value_counts.plot(kind='line',marker='o',color='red')

# Adding labels and title
plt.xlabel('Product Category')
plt.ylabel('Number of Products')
plt.title('Product Count by Category')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Area Plot

plt.figure(figsize=(10,6))
value_counts.plot(kind='area' ,color='orange', alpha=0.5)

plt.xlabel('Product Category')
plt.ylabel('Number of Products')
plt.title('Product Count by Cateory')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Seaborn count plot

plt.figure(figsize=(10,6))
sns.barplot(x=value_counts.index, y=value_counts.values, palette='viridis')

plt.xlabel('Product Category')
plt.ylabel('Number of Products')
plt.title('Product Count by Cateory')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
